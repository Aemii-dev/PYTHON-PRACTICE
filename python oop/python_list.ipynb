```python
# python list
#Lists are used to store multiple items in a single variable
fruit = ["apple", "banana", "cherry"]
print(fruit)
```

```python
#list length
fruit = ["apple", "banana", "cherry"]
print(len(fruit))
```

```python
#Print the second item of the list using by index
fruit = ["apple", "banana", "cherry"]
print(fruit[1])
```

```python
#Change list items
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant"
print(thislist)
```

```python
# e.g:1
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:2] = ["blackcurrant", "watermelon"]
print(thislist)
# e.g:2
marks = [70, 75, 80, 65, 60]

# index 1 se 3 tak marks change karna
marks[1:4] = [85, 90, 88]
print(marks)
# e.g:3
products = ["Shoes", "Bag", "Watch", "Cap", "Belt"]

products[1:3] = ["Laptop Bag", "Smart Watch"]

print(products)
```

```python
# Add list Items
# Append Items
# To add an item to the end of the list, use the append() method:
# e.g:1
thislist = ["apple", "banana", "cherry"]
thislist.append("orange")
print(thislist)
# e.g:2
student =["name", "age", "roll no"]
student.append("id")
print(student)
```

```python
# insert item
# The insert() method inserts an item at the specified index:
# e.g:1
students = ["Ali", "Sara", "Ahmed"]

students.insert(1, "Ayesha")

print(students)
# e.g:2
students =["Ali", "Sara", "Ahmad"]
students.insert(2, "Eman")
print(students)
```

```python
# removed list items
# e.g:1
students = ["Ali", "Sara", "Ahmed"]

students.remove("Sara")

print(students)
# e.g:2
fruits = ["apple", "banana", "cherry"]
fruits.remove("banana")
print(fruits)
```

```python
# Removed Specifie3d Index
# The pop() method removes the specified index.
# e.g:1
fruits = ["apple", "banana", "cherry"]
fruits.pop(1)
print(fruits)
# e.g:2
# If you do not specify the index, the pop() method removes the last item.
fruits =["apple", "banana", "cherry"]
fruits.pop()
print(fruits)
```

```python
# the del keyword also remove the specified list
fruits =["apple", "banana", "cherry"]
del fruits[2]
print(fruits)
# The del keyword can also delete the list completely.
fruits = ["apple", "banana", "cherry"]
del fruits
```

```python
# Loop through a list
# You can loop through the list items by using a for loop
# e.g:1
students = ["Ali", "Sara", "Ahmed", "Ayesha"]

for name in students:
    print(name)
# e.g:2
subjects = ["Math", "Physics", "Computer", "English"]

for subject in subjects:
    print("Subject:", subject)

# e.g:3
prices = [100, 200, 300, 400]

for price in prices:
    print("Price is:", price)

# e.g:4
attendance = ["Present", "Absent", "Present", "Present"]

for status in attendance:
    print(status)
```

```python
# Loop through the index numbers
# Use the range() and len() functions to create a suitable iterable.
# e.g:1
students = ["Ali", "Sara", "Ahmed"]

for i in range(len(students)):
    print(students[i])
fruits = ["apple", "banana", "cherry"]
for i in range(len(fruits)):
  print(fruits[i])
```

```python
# List Comprehension
# List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.
# e.g:1
marks = [50, 60, 70, 80]
new_marks = []
for m in marks:
    new_marks.append(m + 5)
print(new_marks)

# e.g:2
marks = [50, 60, 70, 80]
new_marks = [m + 5 for m in marks]
print(new_marks)

# e.g:3
marks = [35, 60, 45, 80, 30]
pass_students = [m for m in marks if m >= 40]
print(pass_students)

# e.g:4
names = ["ali", "sara", "ahmed"]
upper_names = [n.upper() for n in names]
print(upper_names)
```

```python
# Join two lists
# e.g:1
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list3 = list1 + list2
print(list3)
# append list2 into list1
# e.g:2
list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

for x in list2:
  list1.append(x)

print(list1)
# Using Extend Method
list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

list1.extend(list2)
print(list1)
```

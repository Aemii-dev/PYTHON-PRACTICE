```python
# Python Variables - Assign Multiple Values
# Many Values to Multiple Variables.
# Python allows you to assign values to multiple variables in one line:
# example no 1
name, age, marks = "Ali", 20, 85
print(name)
print(age)
print(marks)
# example no 2
mobile_name, price, color = "Samsung", 50000, "Black"
print(mobile_name)
print(price)
print(color)
```

```python
# One Value to Multiple Variables
# And you can assign the same value to multiple variables in one line:
# e.g:1
boy1 = boy2 = boy3 = "Present"
print(boy1)
print(boy2)
print(boy3)
# e.g:2
name = age = height ="girl"
print(name)
print(age)
print(height)
```

```python
# Global Variables
# Variables that are created outside of a function (as in all of the examples in the previous pages) are known as global variables.

# Global variables can be used by everyone, both inside of functions and outside.
school_name = "City School"   # Global variable

def student():
    print("Student studies in", school_name)

def teacher():
    print("Teacher works in", school_name)

student()
teacher()
```

```python
# e.g2
x = "awesome"

def myfunc():
  x = "fantastic"
  print("Python is " + x)

myfunc()

print("Python is " + x)
```

```python
# e.g3
# If you use the global keyword, the variable belongs to the global scope.
def myfunc():
  global x
  x = "fantastic"

myfunc()

print("Python is " + x)
```

```python
# Python Datatypes built-in Datatypes
# Variables can store data of different types, and different types can do different things
# Student info System
# String (text)
student_name = "Ali"

# Integer (whole number)
student_age = 20

# Float (decimal number)
student_gpa = 3.75

# Boolean (True/False)
is_present = True

# List (multiple values)
subjects = ["Math", "Physics", "Computer"]

# Tuple (fixed values)
semester = (1, 2, 3, 4)

# Dictionary (key-value pair)
student_info = {
    "name": "Ali",
    "age": 20,
    "department": "Computer Science"
}

print("Name:", student_name)
print("Age:", student_age)
print("GPA:", student_gpa)
print("Present:", is_present)
print("Subjects:", subjects)
print("Semesters:", semester)
print("Student Info:", student_info)
```

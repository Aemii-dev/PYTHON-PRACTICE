```python
import pandas as pd

df = pd.read_csv("student_data_clean.csv")
```

```python
print(df)
```

```python
#average study hours by Department
df.groupby('Department')['Study_Hours_Per_Day'].mean()
```

```python
#  Maximum CGPA in Each Department
# Har Department ka highest CGPA find karo.
df.groupby('Department')['CGPA'].max()
```

```python
# Total Students in Each Semester
# Har Semester me kitne students hain?

df['Semester'].value_counts()
```

```python
# create pivot table
pivot_table = pd.pivot_table(df,
                             index="Department",
                             columns="Semester",
                             aggfunc="count")

pivot_table
```

```python
# Pivot table with CGPA mean
pivot_table = pd.pivot_table(df,
                             values="CGPA",
                             index="Department",
                             columns="Semester",
                             aggfunc="mean")

pivot_table
```

```python
# Attendence calculate with semester
pivot_attendance = pd.pivot_table(df,
                                  values="Attendance_Rate",
                                  index="Department",
                                  columns="Gender",
                                  aggfunc='sum')
# percentage mein convert
pivot_attendance = (pivot_attendance * 100).round(0)
pivot_attendance
```

```python
pivot_attendance = pd.pivot_table(df,
                                  values="Attendance_Rate",
                                  index="Department",
                                  columns="Gender",
                                  aggfunc="mean")

# percentage mein convert
pivot_attendance = pivot_attendance * 100

pivot_attendance
```

```python
# Students with Attendance Above 90%
# Un students ka data show karo jinki Attendance_Rate > 90 hai.
df['Attendance_Rate'] > 90
```

```python
# Average Midterm Marks by Gender
# Male vs Female students ke average Midterm_Marks compare karo.
df.groupby('Gender')['Midterm_Marks'].mean()
```

```python
# Average Final Marks of Scholarship Students
df[df['Scholarship']=='Yes']['Final_Marks'].mean()
```

```python
# Count of Students by Gender
df['Gender'].value_counts()
```

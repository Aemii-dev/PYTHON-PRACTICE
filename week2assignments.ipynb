```python
# WEEK 2: ASSIGNMENT 1: The Data Analyzer
# Objective: Create a program that uses functions to perform basic statistical calculations on a list of numbers.

# Instructions:
# 1. Write a function get_numbers() that asks the user to enter numbers separated by spaces and returns them as a list of floats.
# 2. Write a function calculate_mean(data) that returns the mean (average).
# 3. Write a function calculate_variance(data, mean) that calculates sample variance.
# 4. Write a function calculate_std_dev(variance) that returns the square root of the variance.
# 5. In the main program:
#    - Call get_numbers()
#    - Call calculate_mean()
#    - Call calculate_variance()
#    - Call calculate_std_dev()
#    - Print results in formatted output.

# Example Output:
# Enter numbers separated by spaces: 10 12 23 23 16 23 21 16

# --- Statistical Analysis ---
# Data: [10.0, 12.0, 23.0, 23.0, 16.0, 23.0, 21.0, 16.0]
# Mean: 18.0
# Variance: 24.57
# Standard Deviation: 4.96

import math   # for square root

# Function 1: Get numbers from user
def get_numbers():
    nums = input("Enter numbers separated by spaces: ")
    num_list = [float(x) for x in nums.split()]   # convert to float list
    return num_list

# Function 2: Calculate mean
def calculate_mean(data):
    return sum(data) / len(data)

# Function 3: Calculate sample variance
def calculate_variance(data, mean):
    total = 0
    for x in data:
        total += (x - mean) ** 2   # (x - mean)^2
    return total / (len(data) - 1)   # sample variance formula

# Function 4: Calculate standard deviation
def calculate_std_dev(variance):
    return math.sqrt(variance)

# -------- Main Program --------

data = get_numbers()

mean = calculate_mean(data)
variance = calculate_variance(data, mean)
std_dev = calculate_std_dev(variance)

# Print results
print("\n--- Statistical Analysis ---")
print("Data:", data)
print("Mean:", round(mean, 2))
print("Variance:", round(variance, 2))
print("Standard Deviation:", round(std_dev, 2))
```

```python
# ASSIGNMENT 2: The Student Class System
# Objective: Build a simple class system using OOP.
# Instructions:
# 1. Create a base class Person with attributes name and age.
# 2. Create method display_info() that prints name and age.
# 3. Create derived class Student that inherits Person.
# 4. Add attributes student_id and major.
# 5. Override display_info() to print student details.
# 6. Create derived class GraduateStudent inheriting Student.
# 7. Add thesis_topic attribute and override display_info().
# 8. Create one object of each class and call display_info().

# Example Output:
# --- Person ---
# Name: John Doe
# Age: 45
# --- Student ---
# Name: Jane Smith
# Age: 20
# Student ID: S12345
# Major: Computer Science
# --- Graduate Student ---
# Name: Bob Johnson
# Age: 27
# Student ID: G67890
# Major: Physics
# Thesis Topic: Quantum Computing

# Base Class
class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Method to display info
    def display_info(self):
        print("\n--- Person ---")
        print("Name:", self.name)
        print("Age:", self.age)


# Derived Class: Student
class Student(Person):

    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)   # call parent constructor
        self.student_id = student_id
        self.major = major

    # Override display_info()
    def display_info(self):
        print("\n--- Student ---")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)


# Derived Class: GraduateStudent
class GraduateStudent(Student):

    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)
        self.thesis_topic = thesis_topic

    # Override display_info()
    def display_info(self):
        print("\n--- Graduate Student ---")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)
        print("Thesis Topic:", self.thesis_topic)


# -------- Creating Objects --------

# Person object
person1 = Person("ali", 45)

# Student object
student1 = Student("Eman", 20, "S12345", "Computer Science")

# Graduate Student object
grad_student1 = GraduateStudent("Sameer", 27, "G67890", "Physics", "IT Computing")


# -------- Calling Methods --------
person1.display_info()
student1.display_info()
grad_student1.display_info()
```

```python
# ASSIGNMENT 3: Shape Calculator with Magic Methods
# Objective: Demonstrate operator overloading.

# Instructions:
# 1. Create class Rectangle with width and height.
# 2. Implement __str__ method returning Rectangle(width=?, height=?).
# 3. Implement area() returning width * height.
# 4. Overload + operator using __add__ to return new rectangle.
# 5. Overload == operator using __eq__ to compare areas.
# 6. Create rectangles rect1(4,5) and rect2(2,3).
# 7. Print rectangles and areas.
# 8. Create rect3 = rect1 + rect2.
# 9. Compare rectangles.

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    # String representation
    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"

    # Area method
    def area(self):
        return self.width * self.height

    # Overloading + operator
    def __add__(self, other):
        new_width = self.width + other.width
        new_height = self.height + other.height
        return Rectangle(new_width, new_height)

    # Overloading == operator
    def __eq__(self, other):
        return self.area() == other.area()


# Step 6: Create rectangles
rect1 = Rectangle(4, 5)
rect2 = Rectangle(2, 3)

# Step 7: Print rectangles and areas
print(rect1)
print("Area of rect1:", rect1.area())

print(rect2)
print("Area of rect2:", rect2.area())

# Step 8: Add rectangles
rect3 = rect1 + rect2
print("rect3 (after addition):", rect3)
print("Area of rect3:", rect3.area())

# Step 9: Compare rectangles
print("rect1 == rect2:", rect1 == rect2)
```

```python
# Python Fundamentals - Week 1 Assignments & Quizzes
# Assignment 1: The Personal Assistant.
# Objective: Create a simple program that acts as a basic personal assistant by asking for user input and performing simple operations
# Instructions:
# 1. Ask the user for their name and store it in a variable.
# 2. Ask the user for their birth year (as an integer) and store it.
# 3. Ask the user for their current age (as an integer) and store it.
# 4. Calculate and print the year they will turn 100 years old. (Hint: 100 - age + current_year).
# 5. Ask the user for a lucky number (as a float).
# 6. Print a summary message that includes all the information you've collected. Use appropriate data types and typecasting.
# 7. Add comments to your code explaining each step.

# Ask user name
name = input("What is your name? ")
# Ask birth year (convert into integer)
birth_year = int(input("What year were you born? "))
# Ask current age (integer)
age = int(input("What is your current age? "))
# Get current year
current_year = 2026

# Calculate year when user turns 100
year_100 = current_year + (100 - age)
# Ask lucky number (float)
lucky_number = float(input("Enter your lucky number: "))
# Print summary
print("\n--- Personal Assistant Summary ---")

print("Hello,", name + "!")
print("You were born in", birth_year)
print("You are currently", age, "years old")
print("You will turn 100 in the year", year_100)
print("Your lucky number is", lucky_number)
```

```python
# Assignment 2: The Shopping List Manager (Day 4 Data Structures)
# Objective: Build a program that manages a shopping list using lists and dictionaries.

# Instructions:
# 1. Create a list named shopping_list containing at least 5 initial items (e.g., "milk", "bread", "eggs", "apples", "chicken").
# 2. Print the entire list, and also print the third item in the list.
# 3. Add two new items to the list using an appropriate list method.
# 4. Remove one item from the list.
# 5. Create a dictionary named item_prices. The keys should be items from your shopping list (at least 4), and the values should be their prices (as floats).
# 6. Print the price of one specific item from the dictionary.
# 7. Print a formatted summary of your final shopping list and the price dictionary.

# Example Output:
# Initial Shopping List: ['milk', 'bread', 'eggs', 'apples', 'chicken']
# The third item is: eggs
# Adding 'cheese' and 'coffee'...
# Removing 'chicken'...
# Updated Shopping List: ['milk', 'bread', 'eggs', 'apples', 'cheese', 'coffee']
# Item Prices: {'milk': 2.5, 'bread': 1.99, 'eggs': 3.25, 'apples': 0.99}
# The price of bread is $1.99
# --- Final Summary ---
# Items to buy: milk, bread, eggs, apples, cheese, coffee
# Price lookup: {'milk': 2.5, 'bread': 1.99, 'eggs': 3.25, 'apples': 0.99}

# Step 1: Create a shopping list with 5 items
shopping_list = ["milk", "bread", "eggs", "apples", "chicken"]

# Step 2: Print full list and third item
print("Initial Shopping List:", shopping_list)
print("The third item is:", shopping_list[2])   # index 2 = third item

# Step 3: Add two new items
print("\nAdding 'cheese' and 'coffee'...")
shopping_list.append("cheese")
shopping_list.append("coffee")

# Step 4: Remove one item
print("Removing 'chicken'...")
shopping_list.remove("chicken")

# Print updated list
print("Updated Shopping List:", shopping_list)

# Step 5: Create dictionary of item prices
item_prices = {
    "milk": 2.5,
    "bread": 1.99,
    "eggs": 3.25,
    "apples": 0.99
}

# Step 6: Print price of one item
print("\nThe price of bread is $", item_prices["bread"])

# Step 7: Final summary
print("\n--- Final Summary ---")
print("Items to buy:", ", ".join(shopping_list))
print("Price lookup:", item_prices)
```

```python
# Assignment 3: The Number Cruncher (Loops and Conditionals)
# Objective: Write a script that processes a range of numbers using loops and conditional statements.

# Instructions:
# 1. Use a for loop to iterate through numbers from 1 to 20.
# 2. For each number:
#    - If divisible by 3, print "Fizz".
#    - If divisible by 5, print "Buzz".
#    - If divisible by both 3 and 5, print "FizzBuzz".
#    - Otherwise, print the number itself.
# 3. After the loop, ask the user for a number.
# 4. Use a while loop to print a countdown from that number down to 1, and then print "Blast Off!".
# 5. Use break to stop the countdown if the user's number is greater than 10.

# Example Output:
# 1
# 2
# Fizz
# 4
# Buzz
# Fizz
# 7
# 8
# Fizz
# Buzz
# 11
# Fizz
# 13
# 14
# FizzBuzz
# 16
# 17
# Fizz
# 19
# Buzz
# Enter a number for countdown: 5
# Countdown: 5
# Countdown: 4
# Countdown: 3
# Countdown: 2
# Countdown: 1
# Blast Off!
# (If the user enters 12, program ends without countdown due to the break condition)

# Step 1 & 2: FizzBuzz using for loop
for num in range(1, 21):

    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")

    elif num % 3 == 0:
        print("Fizz")

    elif num % 5 == 0:
        print("Buzz")

    else:
        print(num)

# Step 3: Ask user for a number
user_num = int(input("\nEnter a number for countdown: "))

# Step 5: Break condition
if user_num > 10:
    print("Number is greater than 10. Countdown stopped!")

else:
    # Step 4: Countdown using while loop
    while user_num >= 1:
        print("Countdown:", user_num)
        user_num -= 1

    print("Blast Off!")
```

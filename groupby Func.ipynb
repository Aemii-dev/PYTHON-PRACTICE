```python
import pandas as pd

df = pd.read_csv("sales.csv")
df.head()
```

```python
#Display the first 8 rows and check the basic structure of the dataset using .info()
df.head(8)
```

```python
#Basic Structure of dataset
df.info()
```

```python
#Get statistical summary of numerical columns using .describe()
df.describe()
```

```python
#Check Column Name
df.columns

```

```python
#Rename Column
df.rename(columns={"order_id": "OrderID"}, inplace=True)
```

```python
#Data Cleaning Task
#Check Missing Values
df.isnull().sum()
```

```python
#Handle Missing Values
df["price"].fillna(df["price"].mean(), inplace=True)
```

```python
#Check Duplicates
df.duplicated().sum()
```

```python
#Remove Duplicates
df.drop_duplicates(inplace=True)
```

```python
#Convert Date Column to datatime
df["date"] = pd.to_datetime(df["date"])
```

```python
#Checks Datatypes
df.dtypes
```

```python
#Check data types of all columns and convert if necessary (e.g., ensure 'quantity' is int, price in float
df["quantity"] = df["quantity"].astype(int)

df["price"] = df["price"].astype(float)
```

```python
#Standardized Region Values
df["region"] = df["region"].str.strip().str.title()
```

```python
#DATA ANALYSIS TASKS
#Total sales by region
df["sales"] = df["quantity"] * df["price"]
sales_by_region = df.groupby("region")["sales"].sum()

print(sales_by_region)
```

```python
#Region with Highest sales
sales_by_region.idxmax()
```

```python
#Average Discount by product category
df.groupby("category")["discount"].mean()
```

```python
#top five products by quantity sold
df.groupby("product")["quantity"].sum().sort_values(ascending=False).head(5)
```

```python
#Monthly Sales Trend
df["month"] = df["date"].dt.month
df.groupby("month")["sales"].sum()
```

```python
#Discount impact on sales
df.groupby(df["discount"] > 0)["sales"].mean()
```

```python
#Customer Behaviour
df.groupby("customer_type")[["quantity","sales"]].mean()
```
